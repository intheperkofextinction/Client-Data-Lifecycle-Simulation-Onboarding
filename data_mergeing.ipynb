```python
import pandas as pd
```

```python
df_transactions = pd.read_csv(r"C:\Users\amalm\OneDrive\Desktop\finamcial_data_analysis_lerning\project\Client Data Lifecycle Simulation\LI-Small_Trans.csv")
df_accounts = pd.read_csv(r"C:\Users\amalm\OneDrive\Desktop\finamcial_data_analysis_lerning\project\Client Data Lifecycle Simulation\LI-Small_accounts.csv")
```

```python
# Renameing cloumn names for clarity
df_transactions.rename(columns={
    'Account': 'From Account',
    'Account.1' : 'To Account'
},inplace = True)
    
```

```python
# Sender side merge
df_merged = df_transactions.merge(
    df_accounts,
    left_on = ['From Bank', 'From Account'],
    right_on = ['Bank ID', 'Account Number'],
    how='left'
).rename(columns={
    'Entity ID': 'Sender Entity ID',
    'Entity Name': 'Sender Entity Name',
    'Bank Name': 'From Bank Name'
})
```

```python
# Receiver-side merge
df_merged = df_merged.merge(
    df_accounts,
    left_on=['To Bank', 'To Account'],
    right_on=['Bank ID', 'Account Number'],
    how='left',
    suffixes=('', '_Receiver')
).rename(columns={
    'Entity ID': 'Receiver Entity ID',
    'Entity Name': 'Receiver Entity Name',
    'Bank Name': 'To Bank Name'
})
```
